```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
for start in ['Train/Offset_Loss', 'Train/State_Loss', 'Validation/Offset_Loss', 'Validation/State_Loss', 'Validation/State_Accuracy']:
    fig, ax = plt.subplots(figsize=(5, 5))
    for path in ['50_seq', 'base', '150_seq']:
        data = pd.read_csv('./data/' + path + '.csv')
        subset = data[data['metric'].str.startswith(start)]
        if path == 'base':
            path += '/100_seq'
        ax.plot(subset['step']+1, subset['value'], label=path)
    plt.title("Sequence Length Experiment: " + start)
    plt.legend()
    ax.xaxis.set_ticks(np.arange(0, 21, 5))
    plt.show()
```

```python
for start in ['Train/Offset_Loss', 'Train/State_Loss', 'Validation/Offset_Loss', 'Validation/State_Loss', 'Validation/State_Accuracy']:
    fig, ax = plt.subplots(figsize=(5, 5))
    for path in ['64_embed', '128_embed', 'base']:
        data = pd.read_csv('./data/' + path + '.csv')
        subset = data[data['metric'].str.startswith(start)]
        if path == 'base':
            path += '/256_embed'
        ax.plot(subset['step']+1, subset['value'], label=path)
    plt.title("Embedding Dim Experiment: " + start)
    plt.legend()
    ax.xaxis.set_ticks(np.arange(0, 21, 5))
    plt.show()
```

```python
for start in ['Train/Offset_Loss', 'Train/State_Loss', 'Validation/Offset_Loss', 'Validation/State_Loss', 'Validation/State_Accuracy']:
    fig, ax = plt.subplots(figsize=(5, 5))
    for path in ['base', '10_2_2_model', '10_10_1_model', '20_20_1_model']:
        data = pd.read_csv('./data/' + path + '.csv')
        subset = data[data['metric'].str.startswith(start)]
        if path == 'base':
            path += '/4_2_2_model'
        ax.plot(subset['step']+1, subset['value'], label=path)
    plt.title("Model Experiment: " + start)
    plt.legend()
    ax.xaxis.set_ticks(np.arange(0, 21, 5))
    plt.show()
```
